Resizing the image using pillow library

```python
from PIL import Image
```

```python
img = Image.open('/content/dogesh-bhai')
img_resized = img.resize((200,200))
```

```python
img_resized.save('dogesh_bhai_resized.jpg')
```

```python
#displaying image from numpy array
img_res = mpimg.imread('/content/dogesh_bhai_resized.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

Convert RGB Image to greyscale image using OpenCV

```python
#importing opencv library
import cv2
```

```python
img = cv2.imread('/content/dogesh-bhai')
```

```python
type(img)
```

```python
img.shape
```

```python
greyscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(greyscale_image)
```

```python
greyscale_image.shape
```

```python
#saving the greyscale image
cv2.imwrite('dogesh_bhai_greypic.jpg',greyscale_image)
```
